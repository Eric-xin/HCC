# Inference of the model

```python
from matplotlib import pyplot as plt
from utils import Inference
from src.CNN_model import CNNModel
```

```python
model = CNNModel()
model_pth = 'models/CNN_100_epoch_quantized_model.pth'
print(model_pth)
```

```python
from PIL import Image, ImageFilter

test_img = 'data/hydrocarbon_test/test_8.png'
test_img_enhanced = test_img.replace(".png", "_enhanced.png")

test_prediction = Inference(image_paths=[test_img_enhanced], model=model, model_path=model_pth, quantized=True)

# Display the image with the enhanced image and the prediction
fig, axes = plt.subplots(1, 2, figsize=(10, 10), dpi=200)
img = plt.imread(test_img)
axes[0].imshow(img)
# axes[0].axis('off')
axes[0].set_title('Original Image')
axes[1].imshow(Image.open(test_img_enhanced))
# axes[1].axis('off')
axes[1].set_title('Enhanced Image')
plt.suptitle(f'Model Output \n Carbon: {test_prediction[0][0]:.10g} | Hydrogen: {test_prediction[0][1]:.10g}', fontsize=15, y=0.27)
plt.show()
```
